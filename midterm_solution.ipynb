```python
#obtaining the data
import pandas as pd
dataraw = pd.read_csv("https://ies-python-midterm.s3.eu-central-1.amazonaws.com/companies/companies_no_subindustry.csv")
dataraw
```

```python
#finding how many companies have missing date of inclusion
num_missing = dataraw['included'].isna().sum()
print(f'There are {num_missing} companies with missing date of inclusion.')
```

```python
data_fin = dataraw.rename(columns={'GICS Sector': 'GICS'})
print(data_fin)
```

```python
#editing type of data
dataraw.columns = [c.replace(' ', '_') for c in dataraw.columns]
print(dataraw)
dataraw[["Symbol", "Security", "hq", "included", "Founded"]] = dataraw[["Symbol", "Security", "hq", "included", "Founded"]].apply (pd.to_numeric, errors = "coerce")
dataraw["Founded"] = pd.to_datetime (dataraw["Founded"])
dataraw.dtypes
```

```python
#getting rid of zero values
data = dataraw.dropna(subset=['included'])

data['included'] = pd.to_datetime(data['included'], dayfirst=True, errors='coerce')

current_year = pd.Timestamp.now().year
"""data['Symbol'] = current_year - data['Founded']"""

#oldest and youngest
oldest_index = data['Founded'].idxmin()
youngest_index = data['Founded'].idxmax()

print("Oldest Constituent:")
print(data.loc[oldest_index])
print("Youngest Constituent:")
print(data.loc[youngest_index])




```

```python
import matplotlib.pyplot as plt

# Convert the 'included' column to datetime format
dataraw['included'] = pd.to_datetime(data['included'], dayfirst=True, errors='coerce')

# Count the number of companies in each sector
sector_counts = dataraw["GICS_Sector"].value_counts()

# Plot the results as a pie chart
plt.pie(sector_counts, labels=sector_counts.index, autopct='%1.1f%%')
plt.title('Distribution of Companies Across Sectors')
plt.show()
```
